2. Maknanya tidak bisa melakukan operator pembagian karena data X yang diberikan ada string "Cindi" dan "8" yang bersifat tidak dapat dibagi.

```
data = [10,9,"Cindi",4,"8"]
for x in data :
  try :
   print(x / 2)
  except:
    print("Bukan bilangan")
```
